Data dictionary

**Crime ID**: Unique identifier for each individual crime report.
**Month**: The month and year the crime occurred (format: YYYY-MM).

**LSOA code**: Lower Super Output Area (LSOA) code representing a small UK geographic unit.
**LSOA name**: Readable name of the LSOA area.
**Crime type**: Specific classification of the offence (e.g. Robbery, Vehicle crime).
**Last outcome category**: Final outcome of the investigation (e.g. Under investigation, No suspect identified).
**Force**: Police force responsible for the area where the crime was recorded.
**Income Deprivation Decile**: Decile ranking for income deprivation (1 = most deprived, 10 = least).
**Employment Deprivation Decile**: Decile for employment-related deprivation (1 = most deprived).

**Environment Deprivation Decile**: Decile based on environmental factors like pollution and green space.
**Distance to Post Office (km)**: Straight-line distance from LSOA centroid to nearest post office.
**Distance to Primary School (km)**: Distance to the nearest primary school.
**Distance to Supermarket (km)**: Distance to the nearest large supermarket.
**Distance to GP (km)**: Distance to the nearest general practitioner (doctor) clinic.
**Homelessness Rate**: Rate of homelessness per 1,000 households in the area.
**Crime unique ID**: Unique deduplicated identifier for each crime event.
**service_access**: Composite score of access to essential services (higher = worse access).
**crime_count**: Total number of crimes recorded in an LSOA for a given month.
**lagged_crime_count**: Previous monthâ€™s crime count (used for predictive modelling).

### **Bivariate analysis**-- Now, ill look at correlations between crime index/crime count and education/ income/ homelessness/ service access

```python
# Convert necessary columns to numeric
cols = [
    'Crime Deprivation Decile', 
    'crime_count',
    'Income Deprivation Decile', 
    'Education Deprivation Decile', 
    'Environment Deprivation Decile', 
    'service_access', 
    'Homelessness Rate'
]

# Ensure numeric types and drop missing values
df_corr = df[cols].apply(pd.to_numeric, errors='coerce') #sanity check/ safety quip

# Compute correlation matrix
corr_matrix = df_corr.corr()

# Extract correlations with 'crime_index' and sort
crime_index_corr = corr_matrix['Crime Deprivation Decile'].drop('Crime Deprivation Decile').sort_values(ascending=False)

# Display as DataFrame
crime_index_corr_df = crime_index_corr.to_frame(name='Correlation with Crime Index')
print(crime_index_corr_df)
```

**Correlation matrix summary**
- Income Decile (0.48): Lower-income (more income deprived) areas show higher crime index.
- Education Decile (0.38): Worser educated areas show higher crime, typically due to high youth disengagement/ weak social infrastructure or capital.
- Homelessness (0.04): practically no relationship.
- Service access (-0.23): the worser the service access, the lower crime deprivation. This is an unexpected finding. One reason for this could be an issue of under reporting. With the lack of services it's harder to reach out for help/ report crime. Alternately, the 'rural effect' means that service deserts have stronger local ties and thus less crime. 
- Environment Decile (0.24): the more polluted an area, with less recreation and green space, faces elevated crime also. 

**Real estate CTA**

- Prioritise buyer protection in urban/ densely populated areas.

```python
#creating temporary dataframe where each row contains unique data on LSOA, lagged crime, crime, force, and month. 
#Each row contains sum of crime per month
ml_df = df[['LSOA code', 'Month', 'Force', 'crime_count', 'lagged_crime_count']].drop_duplicates()

#creating training and testing set
#training set february 2023-2024, what we will train the model on 
#testing set february 2024-2025, what we will test predictions onto 
train_force = train[train['Force'] == force].dropna(subset=['lagged_crime_count', 'crime_count'])
test_force = test[test['Force'] == force].dropna(subset=['lagged_crime_count', 'crime_count']).copy()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

model_summaries = []
test_frames = []

# Loop over each police force
for force in train['Force'].unique():
    train_force = train[train['Force'] == force].dropna(subset=['lagged_crime_count', 'crime_count'])
    test_force = test[test['Force'] == force].dropna(subset=['lagged_crime_count', 'crime_count']).copy()
    
    # Proceed as before...
    X_train = train_force[['lagged_crime_count']]
    y_train = train_force['crime_count']
    X_test = test_force[['lagged_crime_count']]
    y_test = test_force['crime_count']

    model = LinearRegression()
    model.fit(X_train, y_train)

    test_force['predicted'] = model.predict(X_test)
    test_force['residual'] = y_test - test_force['predicted']

    # Save test set for plotting
    test_frames.append(test_force)

    # Save summary
    model_summaries.append({
        'Force': force,
        'Intercept': model.intercept_,
        'Slope': model.coef_[0],
        'MAE': mean_absolute_error(y_test, test_force['predicted'])
    })
```

```python
#tidy output/ model coefficient summary 
summary_df = pd.DataFrame(model_summaries).sort_values('MAE')
print(summary_df[['Force', 'Intercept', 'Slope', 'MAE']])
```

```python
#faceted residual plots 

combined_test = pd.concat(test_frames)

# Residuals vs Fitted
g1 = sns.FacetGrid(combined_test, col='Force', col_wrap=2, height=4.5)
g1.map_dataframe(sns.scatterplot, x='predicted', y='residual', alpha=0.4, color='red')
g1.set_axis_labels("Fitted values", "Residuals")
g1.set_titles(col_template="{col_name}")
g1.fig.suptitle("Residuals vs Fitted by Force", fontsize=16)
g1.fig.subplots_adjust(top=0.9)
plt.show()

# Actual vs Predicted
g2 = sns.FacetGrid(combined_test, col='Force', col_wrap=2, height=4.5)
g2.map_dataframe(sns.scatterplot, x='crime_count', y='predicted', alpha=0.4)
g2.set_axis_labels("Actual", "Predicted")
g2.set_titles(col_template="{col_name}")
g2.fig.suptitle("Actual vs Predicted Crime Count by Force", fontsize=16)
g2.fig.subplots_adjust(top=0.9)
plt.show()
```
